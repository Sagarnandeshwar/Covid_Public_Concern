```python
import json
import pandas as pd

json_file = "./ALLTweets.json"
```

```python
with open(json_file) as f:
    json_string = json.load(f)

df = pd.DataFrame(json_string['data'])
df = df.drop(['id'], axis=1)
```

```python
df.to_csv(r'./ALLTweets.csv', index = False, header = True)
```
